```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Create and load the dataset into Pandas
df = pd.DataFrame({
    "Name": ["Rahul", "Priya", "Siri", "Anuj"],
    "Marks": [70, 80, 90, 60],
    "Hours Studied": [2, 3, 5, 1]
})

print("Original DataFrame:")
print(df)
```

```python
# 2. Add Performance column
df["Performance"] = df["Marks"].apply(lambda x: "Pass" if x > 65 else "Fail")

print("\nDataFrame with Performance:")
print(df)
```

```python
# 3. Visualize with Matplotlib

# Line Chart → Hours vs Marks
plt.figure(figsize=(6, 4))
plt.plot(df["Hours Studied"], df["Marks"], marker="o", linestyle="-", color="blue")
plt.title("Hours Studied vs Marks")
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.show()

# Scatter Plot → Study vs Marks
plt.figure(figsize=(6, 4))
plt.scatter(df["Hours Studied"], df["Marks"], color="red")
plt.title("Hours Studied vs Marks")
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.show()
```

```python
# 4. Seaborn Barplot → Performance vs Marks
plt.figure(figsize=(6, 4))
sns.barplot(data=df, x="Performance", y="Marks", palette="viridis")
plt.title("Performance vs Marks")
plt.show()
```
