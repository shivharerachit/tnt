```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Employee dataset
df = pd.DataFrame({
    "Name": ["Rahul", "Priya", "Amit", "Anuj"],
    "Age": [25, 30, 28, 35],
    "Department": ["HR", "IT", "Finance", "IT"],
    "Salary": [30000, 50000, 45000, 60000]
})
```

```python
# 1. Barplot → Department vs Salary
plt.figure(figsize=(6, 4))
sns.barplot(data=df, x="Department", y="Salary", palette="Blues")
plt.title("Department vs Salary")
plt.show()
```

```python
# 2. Boxplot → Salary distribution
plt.figure(figsize=(6, 4))
sns.boxplot(y=df["Salary"], color="orange")
plt.title("Salary Distribution")
plt.show()
```

```python
# 3. Heatmap using correlation between Age & Salary
plt.figure(figsize=(5, 4))
corr = df[["Age", "Salary"]].corr()
sns.heatmap(corr, annot=True, cmap="YlGnBu")
plt.title("Correlation Heatmap: Age and Salary")
plt.show()
```
